```python
from google.colab import files
import cv2
import numpy as np
import os

# Upload the images
uploaded_files = files.upload()

# Process the uploaded images
for filename in uploaded_files.keys():
    print(f"Uploaded file: {filename}")

    # Read the image
    image = cv2.imdecode(np.frombuffer(uploaded_files[filename], np.uint8), cv2.IMREAD_COLOR)

    # Resize the image to 50% of its original size
    resized_image = cv2.resize(image, None, fx=0.5, fy=0.5)

    # Save the resized image
    output_filename = f"resized_{filename}"
    cv2.imwrite(output_filename, resized_image)

    print(f"Image {filename} resized and saved as {output_filename}.")

# Remove the uploaded images
for filename in uploaded_files.keys():
    os.remove(filename)

print("Resizing complete!")
```
