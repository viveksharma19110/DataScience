```python
import pandas as pd

# Load the Iris dataset
iris_df = pd.read_csv('Iris.csv')

# Display the first few rows of the dataset
print(iris_df.head())
```

```python
# Check the shape of the dataset (rows, columns)
print(iris_df.shape)

# Get an overview of the columns, data types, and non-null values
print(iris_df.info())

# Summary statistics for numerical columns
print(iris_df.describe())

# Count the number of samples for each species
print(iris_df['Species'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of sepal length vs. sepal width
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_df, hue='Species')
plt.title('Sepal Length vs. Sepal Width')
plt.show()

# Pairwise scatter plot of all features colored by species
sns.pairplot(iris_df, hue='Species')
plt.title('Pairwise Scatter Plot')
plt.show()

# Violin plot of petal length distribution by species
sns.violinplot(x='Species', y='PetalLengthCm', data=iris_df)
plt.title('Petal Length Distribution by Species')
plt.show()

# Box plot of sepal width distribution by species
sns.boxplot(x='Species', y='SepalWidthCm', data=iris_df)
plt.title('Sepal Width Distribution by Species')
plt.show()
```
